Task 6: Another Loop exercise

Simulate 100 random variables drawn from a normal distribution with mean 5 and standard deviation 2
Find the mean and standard deviation of your sample (which is going to be close to your inputs but not exactly equal)
Make a new vector that is the normalised version of part 1
Do this once using a for loop
Do this using the R vectorised language (without a for loop)

```python
import numpy as np

mu, sigma = 5, 2 
s = np.random.normal(mu, sigma, 1000000)

print(s.mean())
print(s.std())

s0=s-s.mean()
print(s0.mean())
print(s0.std())

s1=s0/s0.std()
print(s1.mean())
print(s1.std())
```
