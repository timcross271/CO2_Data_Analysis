Create new dataframe for CO2 Emissions (kt)

Creat new dataframe from that, for countries only, other country groupings and world.

```python
co2_total_df = df.loc[df['Series Name'] == 'CO2 emissions (kt)']
co2_total_df=co2_total_df.reset_index()
co2_total_df.to_csv('CO2 Emissions kt.csv')

afg=co2_total_df.loc[co2_total_df['Country Code'] == 'AFG'].index
zwe=co2_total_df.loc[co2_total_df['Country Code'] == 'ZWE'].index+1
co2_country_df = co2_total_df.iloc[afg[0]:zwe[0],:]

afe=co2_total_df.loc[co2_total_df['Country Code'] == 'AFE'].index
umc=co2_total_df.loc[co2_total_df['Country Code'] == 'UMC'].index+1
co2_other_df = co2_total_df.iloc[afe[0]:umc[0],:]

co2_other_world = co2_total_df.loc[co2_total_df['Country Code'] == 'WLD']

print(co2_other_world)
```

```python
co2_country_df_sort=co2_country_df.drop(['index','Unnamed: 0','Series Code','Country Code','Series Name','2021 [YR2021]','2022 [YR2022]','2023 [YR2023]'],axis='columns')
# print(co2_country_df_sort)
co2_country_df_sort.to_csv('co2_country_sort.csv')


for index, df in enumerate(co2_country_df_sort):
    melt_df = co2_country_df_sort.melt(id_vars=['Country Name'], var_name='Year',value_name=co2_country_df['Series Name'][0]) #Reduce year to one column, add df name as data column header
    melt_df = melt_df.sort_values(by=['Country Name', 'Year']) # Sort by country name and then year
    melt_df=melt_df.reset_index(drop=True) # Re index the new sorted data

melt_df['Year']=melt_df['Year'].str[0:4].astype(int)

print(melt_df)
melt_df.to_csv('melt_df.csv')
```
